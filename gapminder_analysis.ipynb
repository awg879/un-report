```python
(
    gapminder
    .filter(['continent','country','lifeExp','year'])
    .pivot(columns="year", values= 'lifeExp',index=['continent','country'])
)
```

```python
(
    gapminder
    .filter(['continent','country','lifeExp','year'])
    .pivot(columns="year", values= 'lifeExp',index=['continent','country'])
    .reset_index()
    .melt(id_vars=['continent','country'], value_name='lifeExp')
)
```
